### **A.2 Why the dataset is interesting**

The selection of the Spotify Songs dataset was based on several technical factors and strong industry relevance:

**1. Relevance to the digital music industry:**  
That big data now drives decisions in the music industry, rather than just producer instinct.  
As stated by Datta et al. (2018), the adoption of streaming platforms has fundamentally changed how music is consumed and discovered, making Big Data analysis a crucial asset for record labels to understand market behavior.  
[`Changing their tune: How consumers adoption of online streaming affects music consumption and discovery`](https://pubsonline.informs.org/doi/abs/10.1287/mksc.2017.1051)

**2. The Loudness War Phenomenon**  
That there is a historical trend where songs are becoming increasingly louder, and we want to prove this with data.  
This dataset allows for empirical testing of the 'Loudness War' phenomenon, a trend of increasing average volume levels in commercial music recordings that has been extensively documented by Serr et al. (2012).  
[`Measuring the Evolution of Contemporary Western Popular Music`](https://www.nature.com/articles/srep00521)

**3. "Hit Song Science" (Popularity Prediction)**  
That audio features (such as danceability/energy) can be used to predict whether a song will be a hit or not.  
This analysis aligns with the field of study known as Hit Song Science, where researchers like Interiano et al. (2018) use extracted audio features to model and predict the commercial success of a song.  
[`Musical trends and predictability of success in contemporary songs in and out of the top charts`](https://royalsocietypublishing.org/doi/10.1098/rsos.171274)

**4. Richness of Audio Feature Variables**  
That features like 'Valence' and 'Energy' are valid representations of musical emotion.  
This dataset utilizes industry-standard Music Information Retrieval (MIR) features provided by the Spotify API, which include psychoacoustic parameters such as Valence and Energy to quantify the emotional perception of music (Spotify for Developers, 2023).  
[`Get Track's Audio Features`](https://developer.spotify.com/documentation/web-api/reference/get-audio-features)

- Objective: To analyze correlations between variables to understand music production patterns and to validate the "Loudness War" hypothesis using Scatter Plots and Heatmaps.

**4. Data Distribution and Composition**  
- Question: How is the song count distributed across the dataset? Is the dataset balanced?  
- Objective: To provide a general overview of the dataset's composition to ensure that the analysis performed is fair and representative.

**Data Cleaning**

### **C.3 Standardization & Transformation**

**1. Standardization**  
Data standardization is the process of converting data to a common format to enable users to process and analyze it.  
[`Data Standardization`](https://www.sisense.com/glossary/data-standardization/)  
###### **Standardize text columns to lowercase and remove excess spaces*

```python
text_cols = ['track_name', 'track_artist', 'track_album_name', 'playlist_name', 'playlist_genre','playlist_subgenre'] # List of text columns to standardize

for col in text_cols:
    if col in df.columns:
        df[col] = df[col].astype(str).str.lower().str.strip()

print("Text standardization completed (lowercase).")
df[['track_name', 'track_artist', 'track_album_name', 'playlist_name', 'playlist_genre','playlist_subgenre']].head() # Check the results of text standardization on the selected columns
```

**2. Transformation**  
Data transformation is an important part of the data integration process in which raw data is converted into a unified format or structure. Data transformation ensures compatibility with the target system and improves data quality and usability. It is an important aspect of data management practices including data dispute, data analysis, and data warehousing.  
[`Data Transformation`](https://www.ibm.com/id-id/think/topics/data-transformation)

The Transformation stage is carried out to improve the quality of information that can be extracted from the dataset.
- The duration variable is converted from milliseconds to minutes to improve interpretability.
- Party Score, which is an aggregation of the Danceability and Energy features, is used to measure a song's potential in the context of social activities.
- The continuous popularity variable is discretized into ordinal categories (Low, Medium, High) to facilitate proportional comparative analysis between genres.

```python
# Convert duration from milliseconds to minutes
if 'duration_ms' in df.columns:
    df['duration_min'] = df['duration_ms'] / 60000
    df['duration_min'] = df['duration_min'].round(2)
    print("Feature 'duration_min' created.")
    
# Create 'Party Score' (Combination of Danceability & Energy)
if 'danceability' in df.columns and 'energy' in df.columns:
    df['party_score'] = (df['danceability'] + df['energy']) / 2
    print("Feature 'party_score' created.")

# Popularity Category
if 'track_popularity' in df.columns:
    def segment_pop(score):
        if score >= 75: return 'High (Hits)'
        elif score >= 50: return 'Medium'
        else: return 'Low'
    
    df['popularity_category'] = df['track_popularity'].apply(segment_pop)
    print("Feature 'popularity_category' created.")
```

## **D. Exploratory Data Analysis**

### **D.1 Minimum of 3 visualizations**  
**1. Bar Chart Visualization**

```python
# Ready Data (Count & Percentage)
genre_counts = df['playlist_genre'].value_counts() 
total_songs = len(df)

# Create DataFrame for summary table (for display only)
genre_table = pd.DataFrame({
    'Genre': genre_counts.index,
    'Jumlah': genre_counts.values,
    'Persentase': (genre_counts.values / total_songs * 100).round(2)
})

# Create Plot
plt.figure(figsize=(12, 7))
ax = sns.barplot(x='Genre', y='Jumlah', data=genre_table, palette='viridis') # Use the attractive 'viridis' color palette
for i, p in enumerate(ax.patches):
    height = p.get_height() # Get the height of the bar (number of songs)
    pct = genre_table.iloc[i]['Persentase'] # Get the percentage value from our table
    # Write text above the bar
    ax.text(
        p.get_x() + p.get_width() / 2.,  # X Position (center of the bar)
        height + 50,                     # Y Position (slightly above the bar)
        f'{int(height)}\n({pct}%)',      # Text: Count & (Percentage%)
        ha="center", fontsize=11, color='black', fontweight='bold'
    )
plt.title('Music Genre Distribution (Count & Percentage)', fontsize=16) # Plot title
plt.ylabel('Number of Songs', fontsize=12) # Y-axis label
plt.xlabel('Genre', fontsize=12) # X-axis label
plt.ylim(0, max(genre_counts) * 1.15) # Provide empty space above so the text is not cut off
plt.grid(axis='y', linestyle='--', alpha=0.5) # Horizontal grid lines to facilitate reading
plt.show() # Display the plot
```

**2. Boxplot Visualization**

```python
# Sort Genres by Highest Median Tempo
order_by_tempo = df.groupby('playlist_genre')['tempo'].median().sort_values(ascending=False).index # We want the fastest genres on the left

plt.figure(figsize=(12, 7)) # Create Boxplot Visualization
sns.boxplot(
    x='playlist_genre', 
    y='tempo', 
    data=df, 
    order=order_by_tempo,
    palette='cool'
)
# 3. Finishing
plt.title('Distribution of Song Tempo by Genre', fontsize=16)
plt.xlabel('Genre', fontsize=12)
plt.ylabel('Tempo (BPM)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

[1.2 Box Plot Interpretation](#d3-interpretasi-untuk-setiap-output)

**3. Stacked Bar Chart Visualization**

```python
# Create Crosstab
cross_tab = pd.crosstab(df['playlist_genre'], df['popularity_category'])
# Convert to Percentage (Normalize)
cross_tab_pct = cross_tab.div(cross_tab.sum(1), axis=0) * 100

# List Genres by Highest 'High (Hits)' Percentage
target_col = 'High (Hits)' 
if target_col in cross_tab_pct.columns:
    cross_tab_pct = cross_tab_pct.sort_values(by=target_col, ascending=False)

# Create Stacked Bar Chart
desired_order = ['High (Hits)', 'Medium', 'Low']
existing_cols = [c for c in desired_order if c in cross_tab_pct.columns]
plt.figure(figsize=(12, 7))
ax = cross_tab_pct[existing_cols].plot(
    kind='bar', 
    stacked=True, 
    colormap='viridis',
    figsize=(12, 7),
    width=0.8
)
# 5. Decoration
plt.title('Proportion of Song Popularity by Genre', fontsize=16)
plt.xlabel('Genre', fontsize=12)
plt.ylabel('Percentage (%)', fontsize=12)
plt.legend(title='Popularity Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.axhline(50, color='white', linestyle='--', linewidth=1, alpha=0.5)

# Display percentage labels in the middle of bars (Optional, for better visualization)
for c in ax.containers:
    ax.bar_label(c, fmt='%.1f%%', label_type='center', color='white', fontsize=9, weight='bold')

plt.tight_layout()
plt.show()
```

## **E. Summary & Conclusion**

### **E.1 Key Findings**  
Analysis of the Spotify Songs dataset yielded three significant findings related to the popularity dynamics and audio characteristics of modern music:  

**1. Inefficiency Inequality Among Genres and Artists:**  
Descriptive analysis shows asymmetry between production volume and market reception. The EDM genre, despite dominating the quantity of songs in the dataset and having the highest energy intensity (average energy = 0.80), recorded the lowest average popularity compared to other genres. A similar pattern is identified at the artist level, where high productivity (number of releases) is not directly proportional to average popularity, indicating that recommendation algorithms and listener preferences prioritize curation quality over release frequency alone.

**2. Structural Shift in Song Duration (20102020 Trend):**  
Longitudinal analysis reveals a consistent phenomenon of song duration compression. There was a decrease in the average song duration from 4.01 minutes in 2010 to 3.27 minutes in 2020. These findings reflect the music industry's adaptation to the attention economy in the streaming era, where shorter durations have the potential to increase replay value.

**3. Non-Acoustic Popularity Values:**  
Based on correlation matrices and regression analysis, audio technical attributes such as loudness and tempo were found to have negligible correlation with song popularity. Conversely, specific sub-genre classifications (such as post-teen pop and permanent wave) show significantly higher average popularity scores, concluding that demographic segmentation and sociocultural factors of the genre are more valid predictors of success than technical parameters of the audio signal.

### **E.2 Limitations**  
The interpretation of these analysis results needs to consider the following methodological limitations:

**1. Absence of Exogenous Variables:**  
This analysis is limited to internal audio features and basic metadata. Crucial external variables in the formation of virality trendssuch as marketing budgets, social media activity (e.g., TikTok trends), and semantic analysis of lyricsare not included in the dataset, so the variance in popularity cannot be explained holistically.

**2. Platform-Specific Bias:**  
The dataset only represents the consumption behavior of Spotify users. Given that Spotify users tend to be predominantly young, the results of the sub-genre popularity analysis (such as the dominance of post-teen pop) may contain selection bias and do not represent the global music ecosystem as a whole (radio, physical media, or other platforms).

**3. Linearity Assumption:**   
The Pearson correlation approach used in heatmaps only captures linear relationships. Complex non-linear relationships between audio features (e.g., the interaction between slow tempo and high valence in certain genres) may not be detected by the statistical methods currently applied.
